## Tasks

Now write and execute SQL queries to solve the assignment tasks.

### Task 1




##### Display the names of the unique launch sites  in the space mission


```python
#Query BQ for SpaceX Data. 

input_tb_spacex_task_1 = (
'''

SELECT DISTINCT(Launch_Site) FROM `courseracourses-382800.ds_capstone.spacex_db` 

'''
)

#create dataframe from query - SpaceX Data 
df_spacex_task_1 = client.query(input_tb_spacex_task_1, location="us-east1").to_dataframe()
df_spacex_task_1.head()
```


### Task 2


#####  Display 5 records where launch sites begin with the string 'CCA' 


```python
#Query BQ for SpaceX Data. 

input_tb_spacex_task_2 = (
'''

SELECT * FROM `courseracourses-382800.ds_capstone.spacex_db` 
WHERE Launch_Site LIKE 'CCA%'

'''
)

#create dataframe from query - SpaceX Data 
df_spacex_task_2 = client.query(input_tb_spacex_task_2, location="us-east1").to_dataframe()
df_spacex_task_2.head()
```

### Task 3




##### Display the total payload mass carried by boosters launched by NASA (CRS)


```python
#Query BQ for SpaceX Data. 

input_tb_spacex_task_3 = (
'''

SELECT SUM(PAYLOAD_MASS__KG_) AS TOTAL_PAYLOAD_MASS_KG FROM `courseracourses-382800.ds_capstone.spacex_db` 
WHERE Customer LIKE '%NASA (CRS)%'

'''
)

#create dataframe from query - SpaceX Data 
df_spacex_task_3 = client.query(input_tb_spacex_task_3, location="us-east1").to_dataframe()
df_spacex_task_3
```

### Task 4




##### Display average payload mass carried by booster version F9 v1.1


```python
#Query BQ for SpaceX Data. 

input_tb_spacex_task_4 = (
'''

SELECT ROUND(AVG(PAYLOAD_MASS__KG_),2) AS AVG_PAYLOAD_MASS_KG FROM `courseracourses-382800.ds_capstone.spacex_db` 
WHERE Booster_Version LIKE '%F9 v1.1%'

'''
)

#create dataframe from query - SpaceX Data 
df_spacex_task_4 = client.query(input_tb_spacex_task_4, location="us-east1").to_dataframe()
df_spacex_task_4
```

### Task 5

##### List the date when the first successful landing outcome in ground pad was acheived.


_Hint:Use min function_ 


```python
#Query BQ for SpaceX Data. 

input_tb_spacex_task_5 = (
'''

SELECT MIN(Date) AS Date FROM `courseracourses-382800.ds_capstone.spacex_db` 
WHERE Landing__Outcome LIKE '%Success (ground pad)%'

'''
)

#create dataframe from query - SpaceX Data 
df_spacex_task_5 = client.query(input_tb_spacex_task_5, location="us-east1").to_dataframe()
df_spacex_task_5
```

### Task 6

##### List the names of the boosters which have success in drone ship and have payload mass greater than 4000 but less than 6000


```python
#Query BQ for SpaceX Data. 

input_tb_spacex_task_6 = (
'''

SELECT DISTINCT(Booster_Version) AS Booster_Version FROM `courseracourses-382800.ds_capstone.spacex_db` 
WHERE Landing__Outcome LIKE '%Success (drone ship)%'
AND PAYLOAD_MASS__KG_ > 4000 AND PAYLOAD_MASS__KG_ < 6000

'''
)

#create dataframe from query - SpaceX Data 
df_spacex_task_6 = client.query(input_tb_spacex_task_6, location="us-east1").to_dataframe()
df_spacex_task_6
```

### Task 7




##### List the total number of successful and failure mission outcomes


```python
#Query BQ for SpaceX Data. 

input_tb_spacex_task_7 = (
'''

SELECT 

(SELECT Count(Mission_Outcome) FROM `courseracourses-382800.ds_capstone.spacex_db` WHERE Mission_Outcome LIKE '%Success%') as Successful_Mission,
(SELECT Count(Mission_Outcome) FROM `courseracourses-382800.ds_capstone.spacex_db` WHERE Mission_Outcome LIKE '%Failure%') as Failed_Mission

FROM `courseracourses-382800.ds_capstone.spacex_db` 

'''
)

#create dataframe from query - SpaceX Data 
df_spacex_task_7 = client.query(input_tb_spacex_task_7, location="us-east1").to_dataframe()
df_spacex_task_7.head(1)
```

### Task 8



##### List the   names of the booster_versions which have carried the maximum payload mass. Use a subquery


```python
#Query BQ for SpaceX Data. 

input_tb_spacex_task_8 = (
'''

SELECT DISTINCT(Booster_Version) AS Booster_Version FROM `courseracourses-382800.ds_capstone.spacex_db` 
WHERE PAYLOAD_MASS__KG_ = (SELECT MAX(PAYLOAD_MASS__KG_) FROM `courseracourses-382800.ds_capstone.spacex_db`)

'''
)

#create dataframe from query - SpaceX Data 
df_spacex_task_8 = client.query(input_tb_spacex_task_8, location="us-east1").to_dataframe()
df_spacex_task_8
```

### Task 9


##### List the failed landing_outcomes in drone ship, their booster versions, and launch site names for in year 2015


```python
#Query BQ for SpaceX Data. 

input_tb_spacex_task_9 = (
'''
SELECT Landing__Outcome, Booster_Version, Launch_Site, Date
FROM `courseracourses-382800.ds_capstone.spacex_db` 

WHERE CAST(Date AS DATE) BETWEEN CAST('2015-01-01' AS DATE) AND CAST('2015-12-31' AS DATE)

AND Landing__Outcome LIKE '%Failure (drone ship)%'

'''
)

#create dataframe from query - SpaceX Data 
df_spacex_task_9 = client.query(input_tb_spacex_task_9, location="us-east1").to_dataframe()
df_spacex_task_9
```

### Task 10

##### Rank the count of landing outcomes (such as Failure (drone ship) or Success (ground pad)) between the date 2010-06-04 and 2017-03-20, in descending order


```python
#Query BQ for SpaceX Data. 

input_tb_spacex_task_10 = (
'''
SELECT Landing__Outcome, 
COUNT(*) AS Count_Landing_Outcomes

FROM `courseracourses-382800.ds_capstone.spacex_db`

WHERE CAST(Date AS DATE) BETWEEN CAST('2010-06-04' AS DATE) AND CAST('2017-03-20' AS DATE)

GROUP BY Landing__Outcome
ORDER BY count_landing_outcomes DESC

'''
)

#create dataframe from query - SpaceX Data 
df_spacex_task_10 = client.query(input_tb_spacex_task_10, location="us-east1").to_dataframe()
df_spacex_task_10
```
